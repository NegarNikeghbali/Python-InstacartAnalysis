This script contains the following points: 
1.Import df_ords_prods_merge.
2.Creat df subset of the first one million entries.
3.Split the data into groups based on some criteria with groupby().
4.produce a single descriptive statistic for the “order_number” column.
5.produce Multiple Aggregations  for the “order_number” column.
6.Create loyalty Flag..7
8.Export the newly created dataframe as ords_prods_merge in a suitable format.

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

1.Import df_ords_prods_merge.

```python
#import .csv
path=r'E:\Instacart Basket Analysis'
df_ords_prods_merge=pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_merge.pkl'))
```

2.Creat df subset of the first one million entries.

```python
df=df_ords_prods_merge[:1000000]
```

```python
df.head(10)
```

3.Split the data into groups based on some criteria with groupby()

```python
df.groupby('product_name')
```

4.produce a single descriptive statistic for the “order_number” column.

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

5.produce Multiple Aggregations  for the “order_number” column.

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

6.Create loyalty flag.
a)Split the data into groups based on the “user_id” column.
b)Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
c)Create a new column, “max_order,” into which you’ll place the results of your aggregation.
d)creat loyalty flag with loc().

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform("max")
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
