3.Creat if statements.If-Statements with User-Defined Functions.

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

6.Create a new column for the updated labels based on the busiest and slowest days.

```python
# Define a function to assign labels to the busiest and slowest days
def label_busiest_days(day):
    if day in [0, 1]:  # Monday and Tuesday are the busiest days (0 and 1 correspond to Monday and Tuesday)
        return "Busiest days"
    elif day in [3, 4]:  # Thursday and Friday are the slowest days (3 and 4 correspond to Thursday and Friday)
        return "Slowest days"
    else:
        return "Regularly busy"  # Other days are regularly busy
```

```python
# Apply the function to create the new column 'busiest_days'
df['busiest_days'] = df['orders_day_of_week'].apply(label_busiest_days)

# Display the DataFrame to verify the changes
print(df.head())
```

7.Create a new column containing these labels called “busiest_period_of_day.

```python
# Define time periods: "Most orders", "Average orders", and "Fewest orders". Peak hours are typically between 8:00 to 16:00, while the least busy hours are early morning and late evening.

# Define "Most orders" as the hours with the highest order frequency.
most_ords = [8, 9, 10, 11, 12, 13, 14, 15, 16]  # Example: 8:00 to 16:00

# Define least busy hours as the hours with the lowest order frequency.
least_ords = [2, 3, 4, 5, 6, 7, 17, 18, 19]  # Example: 2:00 to 7:00 and 17:00 to 19:00 

# Create a function to assign labels based on the order hour.
def assign_busiest_period(hour):
    if hour in most_ords:
        return "Most orders"
    elif hour in least_ords:
        return "Fewest orders"
    else:
        return "Average orders"
```

```python
# Apply the function to the 'order_hour_of_day' column to create the new column 'busiest_period_of_day'.
df_ords_prods_merge['busiest_period_of_day'] = df_ords_prods_merge['order_hour_of_day'].apply(assign_busiest_period)
```
